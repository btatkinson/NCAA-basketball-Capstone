```python


import os
import time
import requests

import numpy as np 
import pandas as pd

from bs4 import BeautifulSoup
from datetime import datetime, timedelta
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry


```

```python


game_id = 401364899


session = requests.Session()
retry = Retry(connect=3, backoff_factor=0.5)
adapter = HTTPAdapter(max_retries=retry)
session.mount('http://', adapter)
session.mount('https://', adapter)

url = f'https://www.espn.com/mens-college-basketball/game/_/gameId/{401364899}'

page = requests.get(url)
soup = BeautifulSoup(page.content, "lxml")
```

```python


soup


```

```python


def extract_game_info(soup):
    
    game_info = soup.find('article', 'game-information')
    location = game_info.find('div','caption-wrapper').text.strip()
    location_detail = game_info.find('div','game-location').text.strip()
    line = game_info.find('div','odds')
    line, ou = line.find_all('li')
    line, ou = line.text, ou.text
    other = game_info.find_all('div','game-info-note','capacity')

    attendance = other[0].text
    capacity = other[1].text
    referees = other[2].text
    
    
    return [location, location_detail, line, ou, attendance, capacity, referees]



location, location_detail, line, ou, attendance, capacity, referees = extract_game_info(soup)


```
