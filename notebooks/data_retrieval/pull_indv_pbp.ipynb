```python

def extract_player_table(soup):
    
    player_blocks = soup.find_all('ul','playerfilter')
    t1_players = player_blocks[0].find_all('li')
    t2_players = player_blocks[1].find_all('li')

    player_table = []
    for p in t1_players:
        player_table.append([p['data-homeaway'], p['data-playerid'], p.text])
    for p in t2_players:
        player_table.append([p['data-homeaway'], p['data-playerid'], p.text])

    player_table = pd.DataFrame(player_table, columns=['side','player_id','player_name'])
    player_table = player_table.loc[player_table['player_id']!="0"].reset_index(drop=True)
    player_table['player_id'] = player_table['player_id'].astype(int)

    return player_table



```
