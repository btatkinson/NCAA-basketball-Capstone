```python

import os
import pandas as pd

DATA_PATH = '../../data/'

```

```python

season = 2022
schedule = pd.read_csv(os.path.join(DATA_PATH, f'ESPN/schedules/by_season/{season}.csv'))
```

```python

print(list(schedule))
```

```python


conf_ids = schedule.copy().groupby(['home.displayName'])['home.conferenceId'].apply(pd.Series.mode).reset_index().drop(columns=['level_1']).sort_values(by='home.conferenceId')
conf_ids['home.conferenceId']= conf_ids['home.conferenceId'].astype(int)
```

```python

team_colors = schedule.copy().groupby(['home.displayName'])['home.color'].apply(pd.Series.mode).reset_index().drop(columns=['level_1']).sort_values(by='home.color')
team_alt_colors = schedule.copy().groupby(['home.displayName'])['home.alternateColor'].apply(pd.Series.mode).reset_index().drop(columns=['level_1']).sort_values(by='home.alternateColor')
```

```python


conference_id2name = {
    1:'American East',
    2:'ACC',
    3:'A10',
    4:'Big East',
    5:'Big Sky',
    6:'Big South',
    7:'Big 10',
    8:'Big 12',
    9:'Big West',
    10:'CAA',
    11:'Conference USA',
    12:'Ivy League',
    13:'MAAC',
    14:'MAC',
    16:'MEAC',
    18:'MVC',
    19:'Northeast',
    20:'OVC',
    21:'PAC-12',
    22:'Patriot League',
    23:'SEC',
    24:'Southern',
    25:'Southland',
    26:'SWAC',
    27:'Sun Belt',
    29:'WCC',
    30:'WAC',
    44:'MWC',
    45:'Horizon',
    46:'ASUN',
    49:'Summit League',
    62:'AAC'
}
```

```python
conf_ids['conference_name'] = conf_ids['home.conferenceId'].map(conference_id2name)
```

```python
conf_ids = conf_ids.merge(team_colors, on=['home.displayName'], how='left')
conf_ids = conf_ids.merge(team_alt_colors, on=['home.displayName'], how='left')
```

```python

conf_ids = conf_ids.sort_values(by=['home.displayName']).reset_index(drop=True)
conf_ids.columns=['team_name','ESPN_conference_id','conference_name','primary_color','secondary_color']
```

```python


conf_ids.to_csv('team_meta.csv', index=False)
```
